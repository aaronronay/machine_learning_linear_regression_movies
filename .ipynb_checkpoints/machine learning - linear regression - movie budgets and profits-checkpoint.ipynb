```python
# Aaron Ronay
# Aug  6, 2019

# linear regression - movie budgets and profits
#
# question: How much money will a movie production make?
# feature, independet variable, X: movie production budget
# target, dependent variable, y: movie profit
#
# workflow steps:
# 1 get data
# 2 clean data
# 3 visualize data
# 4 run a machine learning algorithm
# 5 evaluate results
# 6 make a prediction
#
#
#
#dependencies
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import warnings
#
#
#
# Suppress Warning
%matplotlib inline
warnings.filterwarnings(action="ignore", module="scipy", message="^internal gelsd")
#
#
#
# get data
# obtained from: https://www.the-numbers.com/movie/budgets/all
# date obtained: May 25, 2019
data = pd.read_csv('data_dirty.csv')
data = pd.DataFrame(data)
data.head(10)
```

```python
# clean data - drop columns
data = data.drop('Rank', axis=1)
data.head(10)
```

```python
# clean data - drop columns
data = data.drop('Release Date', axis=1)
data.head(5)
```

```python
# clean data - drop columns
data = data.drop('Domestic Gross ($)', axis=1)
data.head(5)
```

```python
# clean data - rename columns so it's easier to read and work with
data = data.rename(columns={"Movie Title": "Movie", "Production Budget ($)": "Budget", "Worldwide Gross ($)": "Profit"})
data.head(10)
```

```python
# clean data - convert any currency numbers where profit was $0 to a 0
# so that we can later remove any missing values of 0 from our data
data = data.replace(to_replace='$0', value=0)
data.head(10)
```

```python
# clean data - remove any rows where there was a 0 profit or 0 budget
data = data[(data != 0).all(1)]
data.head(10)
```

```python
# clean data - remove commas from currency numbers, then convert currency numbers to regular integers
data[data.columns[1:]] = data[data.columns[1:]].apply(lambda x: x.str[1:].str.split(',').str.join('')).astype(np.int64)
data.head(10)
```

```python
# clean data - reset the index for all the rows
data = data.reset_index(drop=True)
data.head(10)
```

```python
# visualize data - do quick histogram of the data to look at each column's distribution
data.hist()
```

```python
# visualize data - create an X and Y variable for each feature
# and since they exist as 1 dimensional series within a dataframe,
# so put each feature into it's own dataframe
X = pd.DataFrame(data['Budget'])
y = pd.DataFrame(data['Profit'])
```

```python
# visualize data - scatter plot
plt.figure(figsize=(10,6))
# use some transparancy to show overlapping data points, so alpha=0.3
plt.scatter(X, y, alpha=0.3)
plt.title('movie budgets vs movie profits')
plt.xlabel('budget')
plt.ylabel('profit')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```

```python
# run a machine learning algorithm 
# by defining the linear regression model and plug in X and Y
# so, the equation of the linear regression line is as follows:
# y = mx + b ...
# so, which line is best for all the data points?
# we need to minimize the residual values (gaps) between the regression line's fitted (predicted) value, 
# and the actual value for any given data point
# so we choose a line with the smallest residuals
# which is the difference between the actual value and the predicted value, for every data point
# so we then have to square each of these diffences to obtain all positive values,
# then take the sum of all the squared residuals
# so to find the ideal fit for the regression line, we need to choose an
# intercept and slope that minimizes the sum of all the squared residuals
# and this number is called the Residual Sum of Squares
# so, this is where our algorithim will come in.
# we just need to plug in X and y
#
regr = LinearRegression()
model = regr.fit(X, y)
#
# now, this algorithm has done all calculations for 
# all of the tedious possibilities for summing all of the squared 
# differences between the actual and predicted values
# and has created a linear regression model using our X and Y values
# so now, we can
# View the Y intercept
print('The Y intercept is:')
print(model.intercept_)
# view the slope coefficient, theta_1
print('The Slope coefficient is:')
print(model.coef_)
```

```python
# run a machine learning algorithm 
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
# plot the regression line by using X values, and for Y values,
# use model.predict(X), since Y is a function dependent on the X values.
plt.plot(X, model.predict(X), color='red', linewidth=4)
plt.title('movie budgets vs movie profits')
plt.xlabel('budget')
plt.ylabel('profit')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
# evaluate results and make a prediction
# how do we know this is accurate, or how good our model is?
# Let's now examine R^2, goodness of fit
print('the R^2 value is:')
print(model.score(X, y))
print('this value is the amount of variation in movie profits, explained by the movies budget')
print('therefore this linear model, with a single feature of a movie budget, can explain around 55% of the variation in movie profits, which is actually good.')
print('this is a very simplistic model, so futher investigation would be required to determine what other features, if any, affect the profits, and if there are any non-linear relationships')
#
# so, 
# y = mx + c ...
# or
# y = ϴ0 + ϴ1(X)
# or
# hϴ(x) = ϴ0 + ϴ1(X)
# which is equal to :  
# hϴ(x) = -7236192.7  +  3.1115(x)
# where 
# ϴ0 = -7.2 Million
# ϴ1 = 3.1115
# now, let's predict the profit for a movie with a budget of $50 Million:
# hϴ(x) = -7236192.7  +  3.1115(50000000)
# hϴ(x) = 148 Million
```
